```python
# import packages
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
import plotly.express as px
```

# DB_Scan

```python
tr_pi_bf_ff_data = pd.read_csv("tr_pi_bf_data.csv")
tr_pi_bf_ff_data
```

```python
index_df = tr_pi_bf_ff_data.set_index("Location")
index_df
```

```python
index_df.T.isna().sum()
```

```python
clustering = MeanShift(bandwidth=40).fit(index_df)
clustering.labels_
```

```python
# Get cluster labels and predicted centroids
cluster_labels = clustering.labels_
centroids = clustering.cluster_centers_
```

```python
cluster_labels
```

```python
cluster_df = pd.DataFrame({"Location":index_df.index,"K_Means_labels":cluster_labels})
cluster_df
```

```python
cluster_df.to_csv("cluster_df.csv",index=False)
```
